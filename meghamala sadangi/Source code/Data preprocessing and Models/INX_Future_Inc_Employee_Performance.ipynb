### Handle the outliers
To handle outliers, we employ two methods: \
a. IQR method - used when data is not normally distributed.\
b. 3-sigma rule - used when data is normally distributed.\
Since the data is not normally distributed, we are using IQR method to handle the outliers.

```python
for column in data1:
    Q3=data[column].quantile(0.75)
    Q1=data[column].quantile(0.25)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    print(column)
    print(lower_limit,upper_limit)
```

# Feature transformation

## Checking the Distribution of numerical data

```python
# finding Skewness and Kurtosis for numerical columns
Not_fairly_symmetrical=[]
less_peaked=[]
high_peaked=[]
for i in data3.drop('ExperienceYearsAtThisCompany',axis=1):
    print(i,'Skewness:',data[i].skew())
    if data[i].skew()>=-0.5 and data[i].skew()<=0.5:
        print(i,'fairly symmetrical')
    else:
        print(i,'Not fairly symmetrical')
        Not_fairly_symmetrical.append(i)
    print(i,'Kurtosis:',data[i].kurt())
    if data[i].kurt()==3:
        print(i,'Normally distributed')
    elif data[i].kurt()<3:
        print(i,'less peaked')
        less_peaked.append(i)
    else:
        print(i,'high peaked')
        high_peaked.append(i)
```

* If skewness is between range -0.5 and 0.5, therefore the data is fairly symmetrical. If skewness between -1 and -0.5, or between 0.5 and 1, therefore the data is moderately skewed.
* If skewness is greater than one or less than -1, the data is highly skewed.
* If Kurtosis=3, Normally distributed.
* If Kurtosis<3, less peaked.
* If Kurtosis>3, high peaked.

```python
# printing not fairly symmetrical columns
Not_fairly_symmetrical
```

```python
# printing less peaked columns
less_peaked
```

## Hyperparameter tuning for Random Forest Classifier

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False] 
```

```python
#dictionary for hyperparameters
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_clf1 = RandomForestClassifier(random_state=42)#model

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='accuracy',param_distributions=random_grid, n_iter=100, cv=3, 
                               verbose=2, random_state=42, n_jobs=-1)
rf_cv.fit(x_smote, y_smote)
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")
```

```python
# Training the model based on best parameter
rf_clf2 = RandomForestClassifier(n_estimators= 1800, min_samples_split=2, min_samples_leaf= 1, max_features='auto', max_depth=20, bootstrap=False)#passing best parameter to randomforest
rf_clf2.fit(x_smote,y_smote)
y_predict=rf_clf2.predict(x_test)
```
